```python
# importar las librerías necesarias
from sklearn import datasets
from sklearn import svm
from sklearn.model_selection import train_test_split

# cargar la base de datos iris
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 11)
```

```python
# Crear el modelo SVM con kernel lineal
clf_linear = svm.SVC(kernel='linear', C=1)

# Crear el modelo SVM con kernel RBF
clf_rbf = svm.SVC(kernel='rbf', C=1)

# Entrenar ambos modelos con los datos de entrenamiento
clf_linear.fit(X_train, y_train)
clf_rbf.fit(X_train, y_train)
```

```python
# Realizar predicciones con los datos de prueba
y_pred_linear = clf_linear.predict(X_test)
y_pred_rbf = clf_rbf.predict(X_test)

# Calcular la precisión de ambos modelos
accuracy_linear = clf_linear.score(X_test, y_test)
accuracy_rbf = clf_rbf.score(X_test, y_test)
print("Precisión con kernel lineal: ", accuracy_linear)
print("Precisión con kernel RBF: ", accuracy_rbf)
```
