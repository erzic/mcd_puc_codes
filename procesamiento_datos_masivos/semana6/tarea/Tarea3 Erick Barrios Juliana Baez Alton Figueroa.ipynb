# Procesamiento de Datos

```python
with open('canastas_tarea_3.csv', 'r', encoding='utf-8') as f:
    texto = f.read()
```

```python
compras = texto.split("\n")
compras = [tuple(c.split(",")) for c in compras]
print(len(compras))
compras[:10]
```

# 1. Impresión de las 10 reglas de asociación más relevantes:

Escribe un programa que imprima en pantalla las 10 reglas de asociación con mayor soporte, siempre   y cuando tengan una confianza mayor a 0.1

```python
itemsets, rules = apriori(compras, min_support=0.01, min_confidence=0.1) 
```

```python
len(rules)
```

```python
print(rules[0])
rules[0].support
```

```python
regla_tmp = (rules[0].lhs[0], rules[0].rhs[0])
calculo_soporte_manual = sum([1 for c in compras if set(regla_tmp).issubset(set(c))])/len(compras)
if calculo_soporte_manual==rules[0].support:
    print("Los calculos cuadran")
```

```python
rules[0]
```

```python
top_10 = pd.DataFrame({
    'rules':    pd.Series(rules),
    'confianza':pd.Series([i.confidence for i in rules]),
    'soporte':  pd.Series([i.support for i in rules])
}).sort_values(by='soporte', ascending=False).head(10)
```

```python
top_10
```

# 2. Cálculo del interés de las reglas
Define el interés de una regla I -> j como la diferencia (S – F) entre el soporte de la regla (S) y la proporción (F) de canastas que contienen j en relación al total de canastas. Escribe un programa que calcule el interés de las 10 reglas que se encontraron en el punto 1-.

```python
def presencia_canastas(x, canastas):
    return sum([1 for i in canastas if x in i])/len(compras)
```

```python
top_10['presencia_j'] = top_10['rules'].apply(lambda x: presencia_canastas(x=x.lhs[0], canastas=compras))
top_10['interés'] = top_10['soporte'] - top_10['presencia_j']
```
