```python
import pandas as pd
```

```python
hospitales = pd.read_excel('Datos.xls',sheet_name='Hospitales')
historico = pd.read_excel('Datos.xls',sheet_name='Información Histórica')
grd = pd.read_excel('Datos.xls',sheet_name='GRD')
```

## Pregunta 1

```python
servicios_de_salud = pd.merge(
    left=historico[['Hospital al que llega', 'Día de llegada']],
    right=hospitales[['Nombre Establecimiento', 'Nombre Servicio de Salud']],
    left_on='Hospital al que llega',
    right_on='Nombre Establecimiento'
)

llegadas_dia_1 = servicios_de_salud[servicios_de_salud["Día de llegada"]==1]
llegadas_dia_1_hospitales = llegadas_dia_1[['Nombre Servicio de Salud']].value_counts()
llegadas_dia_1_hospitales = llegadas_dia_1_hospitales.reset_index()
llegadas_dia_1_hospitales
```

## Pregunta 3

```python
servicios_de_salud = pd.merge(
    left=historico[['Hospital al que llega', 'Día de llegada', 'GRD']],
    right=hospitales[['Nombre Establecimiento', 'Nombre Servicio de Salud']],
    left_on='Hospital al que llega',
    right_on='Nombre Establecimiento'
)

llegadas_por_hospital = servicios_de_salud[['Nombre Servicio de Salud', 'Hospital al que llega']].value_counts().reset_index()

mediana_de_llegadas_por_hospital = llegadas_por_hospital[['Nombre Servicio de Salud', 'count']].groupby('Nombre Servicio de Salud').median()
mediana_de_llegadas_por_hospital.sort_values(by='count', ascending=False)
```

## Pregunta 5

```python
costo_por_hospital = pd.merge(
    left=historico,
    right=hospitales,
    left_on='Hospital al que llega',
    right_on='Nombre Establecimiento'
)



costo_por_hospital = costo_por_hospital[(costo_por_hospital['Día de llegada'].between(415, 424)) & (costo_por_hospital['Hospital al que llega']=='Hospital el Carmen')]
costo_por_hospital['costo_paciente'] =  costo_por_hospital.apply(lambda x: x[f'Costo diario {x['GRD']} [MM$]']*x['Estadia'], axis=1)

costo_por_hospital['costo_paciente'].sum()
```

## Pregunta 6

```python
costo_por_hospital = pd.merge(
    left=historico,
    right=hospitales,
    left_on='Hospital al que llega',
    right_on='Nombre Establecimiento',
    how='left'
)


costo_por_hospital['costo_total_paciente'] =  costo_por_hospital.apply(lambda x: x[f'Costo diario {x['GRD']} [MM$]']*x['Estadia'], axis=1)
costo_por_hospital['costo_diario'] =  costo_por_hospital.apply(lambda x: x[f'Costo diario {x['GRD']} [MM$]'], axis=1)
costo_por_hospital = costo_por_hospital[costo_por_hospital['Día de llegada']>414]
costos = costo_por_hospital[['Hospital al que llega','GRD', 'costo_diario']].drop_duplicates()

costos = pd.merge(
    left = costos,
    right = costo_por_hospital[['Nombre Servicio de Salud', 'Hospital al que llega']],
    left_on = 'Hospital al que llega',
    right_on = 'Hospital al que llega'
)

costos_promedio = costos.groupby('Nombre Servicio de Salud')['costo_diario'].mean().reset_index()
costos_promedio.columns = ['Nombre Servicio de Salud', 'costo_promedio']
```

```python
union_costos = pd.merge(
    left = costo_por_hospital,
    right = costos_promedio,
    left_on='Nombre Servicio de Salud',
    right_on='Nombre Servicio de Salud'
)

union_costos['total'] = union_costos['costo_promedio']*union_costos['Estadia']


union_costos[union_costos['Día de llegada']>414].groupby('Nombre Servicio de Salud')['total'].sum()
```
