# Pregunta 4
¿Cuál de las siguientes alternativas corresponde a la estimación de la cantidad de pacientes con GRD9 que llegarán el día 425 al Hospital Salvador utilizando una regresión lineal simple sobre los últimos 5 días (días 420 a 424) redondeado al entero superior?

```python
from sklearn.linear_model import LinearRegression

grd9 = historico.loc[(historico['GRD']=='GRD9') & (historico['Día de llegada']>=420) & (historico['Hospital al que llega']=='Hospital Del Salvador')]
grd9 = grd9.groupby('Día de llegada')['Número de llegada'].count()
X = grd9.index.values.reshape(-1,1)
y = grd9.values

regressor = LinearRegression().fit(X,y)

prediccion_dia_425 = regressor.intercept_ + regressor.coef_*425
prediccion_dia_425
```

# Pregunta 5
¿Cuál de las siguientes alternativas corresponde a la estimación de la cantidad de pacientes que llegarán el día 425 al Hospital Salvador que se obtiene utilizando una regresión lineal simple sobre los últimos 5 días (días 420 a 424)? Nota: considere el resultado redondeado al entero más cercano.

```python
from sklearn.linear_model import LinearRegression

grd9 = historico.loc[(historico['Día de llegada']>=420) & (historico['Hospital al que llega']=='Hospital Del Salvador')]
grd9 = grd9.groupby('Día de llegada')['Número de llegada'].count()
X = grd9.index.values.reshape(-1,1)
y = grd9.values

regressor = LinearRegression().fit(X,y)

regressor.predict([[425]])
```

# Pregunta 7
¿Cuál de las siguientes alternativas corresponde al valor del coeficiente R2 (R cuadrado) de una estimación sobre la cantidad de pacientes que llegarán a la red el día 425 utilizando una regresión lineal simple sobre los últimos 5 días (días 420 a 424)?  

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

grd9 = historico.loc[(historico['Día de llegada']>=420)]
grd9 = grd9.groupby('Día de llegada')['Número de llegada'].count()
X = grd9.index.values.reshape(-1,1)
y = grd9.values

regressor = LinearRegression().fit(X,y)
y_pred = regressor.predict(X)

r2_score(y, y_pred)
```

# Pregunta 9
Los valores del intercepto y la pendiente de un modelo de regresión lineal para predecir la cantidad de pacientes que llegarán un día cualquiera al Hospital San José son 60 y 10 respectivamente. La variable independiente de esta regresión corresponde a las llegadas del hace 7 días. 

¿Cuál de las siguientes alternativas corresponde a la estimación de esta regresión para los días 422, 423 y 424?

```python
intercepto = 60
coef = 10
grd9 = historico.loc[(historico['Hospital al que llega']=='Complejo Hospitalario San José')]
grd9 = grd9.groupby('Día de llegada')['Número de llegada'].count()

for i in [grd9[422-7],grd9[423-7],grd9[424-7]]:
    print(f"Prediccion para el dia {i}: {intercepto + i*coef}")
```
