# Pregunta 1:
What was the Annualized Return of the Lo 20 portfolio over the entire period?

```python
p1_returns = datos['Lo 20']/100 # divimos por 100 por temas de formato
n_months = p1_returns.shape[0] # obtenemos el numero de meses o periodos
p1_month = (p1_returns + 1).prod()**(1/n_months) - 1 # componemos los retornos: (1 + R1)(1 + R2)...(1+Rn) para obtener el retorno combinado
# luego se eleva a 1/n_months para mensualizar el retorno combinado y le sustraemos 1 para obtener el % de retorno como tal
p1_annual = (p1_month + 1)**12-1 # anualizamos retorno mensualizado
p1_annual
```

# Pregunta 9 y 10
What was the Max Drawdown (expressed as a positive number) experienced over the 1999-2015 period in the SmallCap (Lo 20) portfolio?

Question 10
At the end of which month over the period 1999-2015 did that maximum drawdown on the SmallCap (Lo 20) portfolio occur? 

```python
def drawdown(return_series: pd.Series):
    """Takes a time series of asset returns.
       returns a DataFrame with columns for
       the wealth index, 
       the previous peaks, and 
       the percentage drawdown
    """
    wealth_index = 1000*(1+return_series).cumprod()
    previous_peaks = wealth_index.cummax()
    drawdowns = (wealth_index - previous_peaks)/previous_peaks
    return pd.DataFrame({"Wealth": wealth_index, 
                         "Previous Peak": previous_peaks, 
                         "Drawdown": drawdowns}).sort_values(by='Drawdown', ascending=True)
```

```python
p9_returns = datos.loc[199901:201599,'Lo 20']/100
p9_returns.index = pd.to_datetime(p9_returns.index, format="%Y%m").to_period('M')

drawdown(p9_returns)
```

```python
datos = pd.read_csv("data/edhec-hedgefundindices.csv",
                      header=0, index_col=0, parse_dates=True)
datos = datos/100
datos.index = datos.index.to_period('M')
datos
```

```python
def semideviation(r):
    """
    Returns the semideviation aka negative semideviation of r
    r must be a Series or a DataFrame, else raises a TypeError
    """
    is_negative = r < 0
    return r[is_negative].std(ddof=0)
```

```python
li = pd.Timestamp(year=2009, month=1, day=1).to_period('M')
ls = pd.Timestamp(year=2018, month=12, day=31).to_period('M')
```

```python
datos.loc[li:ls,:]
```

```python
semideviation(p13_returns).sort_values()
```

# Pregunta 15
Looking at the data since 2009 (including all of 2009) which Hedge Fund Index has been most negatively skewed?   

```python
def skewness(r):
    """
    Alternative to scipy.stats.skew()
    Computes the skewness of the supplied Series or DataFrame
    Returns a float or a Series
    """
    demeaned_r = r - r.mean()
    # use the population standard deviation, so set dof=0
    sigma_r = r.std(ddof=0)
    exp = (demeaned_r**3).mean()
    return exp/sigma_r**3
```

```python
skewness(p15_returns).sort_values()
```

# Pregunta 16
Looking at the data since 2000 (including all of 2000) through 2018 which Hedge Fund Index has exhibited the highest kurtosis?   

```python
def kurtosis(r):
    """
    Alternative to scipy.stats.kurtosis()
    Computes the kurtosis of the supplied Series or DataFrame
    Returns a float or a Series
    """
    demeaned_r = r - r.mean()
    # use the population standard deviation, so set dof=0
    sigma_r = r.std(ddof=0)
    exp = (demeaned_r**4).mean()
    return exp/sigma_r**4
```

```python
p16_returns = datos.loc[pd.Timestamp(year=2000, month=1, day=1).to_period('M'):,:]
p16_returns
```

```python
kurtosis(p16_returns).sort_values()
```
